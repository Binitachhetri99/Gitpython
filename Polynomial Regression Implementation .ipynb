```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = 6*np.random.rand(100,1) -3
y = 0.5 * x**2 + 1.5 * x + 2 + np.random.randn(100,1)
plt.scatter(x, y, color = 'green')

plt.xlabel('x dataset')
plt.ylabel('y dataset')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
## Lets implement a simple linear regression 
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
```

```python
regr.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, regr.predict(x_test))
print(score)
```

```python
## lets visualixe tgis model
plt.plot(x_train, regr.predict(x_train), color = 'r')
plt.scatter(x_train, y_train)
plt.xlabel('X dataset')
plt.ylabel('Y dataset')
```

```python
#increase a degree h(theta) = bo +b1x1 + b2x1^2
# lets apply polynomial transformation 
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2, include_bias = True)
x_train_poly = poly.fit_transform(x_train) # makes sure that the main data
x_test_poly = poly.transform(x_test) #dont wanna know this
```

```python
x_train_poly
```

```python

regression = LinearRegression()
regression.fit(x_train_poly, y_train)
y_pred = regression.predict(x_test_poly)
score = r2_score(y_test,y_pred)
print(score)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
plt.scatter(x_train, regression.predict(x_train_poly))
plt.scatter(x_train, y_train)
```

```python
#3 prediction of new data set 
x_new = np.linspace(-3, 3, 200).reshape(200,1)
x_new_poly = poly.transform(x_new)
x_new_poly
```

```python
y_new = regression.predict(x_new_poly)
plt.plot(x_new, y_new, 'r-', linewidth = 2, label = 'New predictions')
plt.plot(x_train, y_train, 'b.', label = 'Training points')
plt.plot(x_test, y_test, 'g.', label = 'Testing points')
plt.xlabel("x")
plt.ylabel('y')
plt.legend()
plt.show()
```

## Pipeline Concepts 
1. It is series of data processing steps or operations that are chained together sequentially
2. A typical machine learning pipeline involves data preprocessing, feature enigineering, model building. 

```python
#
from sklearn.pipeline import Pipeline
```

```python
def poly_regression(degree):
    x_new = np.linspace(-3, 3, 200).reshape(200,1)
    x_new_poly = poly.transform(x_new)
    poly_features = PolynomialFeatures(degree=degree, include_bias = True)
    lin_reg = LinearRegression()
    poly_regression = Pipeline([
        ("poly_features", poly_features),
        ('lin_reg', lin_reg)
    ])
    poly_regression.fit(x_train, y_train)
    y_pred_new = poly_regression.predict(x_new)
    # plotting prediction line 
    plt.plot(x_new, y_pred_new, 'r', label = 'Degree ' + str(degree), linewidth = 2)
    plt.plot(x_train, y_train, 'b', linewidth = 3)
    plt.plot(x_test, y_test, 'g', linewidth = 3)
    plt.legend(loc = 'upper left')
    plt.xlabel("x")
    plt.ylabel('y')
    plt.axis(-4, 4, 0, 10)
    plt.show()
    
    
```

```python
poly_regression(4)
```
